Now I am going to map some stuff! Below I introduced the simple folium basemap that we were taught in the lab and layered the Total Population that does not carry health insurance for the census tracts I selected for in my dataset. The second map looks at the total number of people WITH health insurance.  

```python
import folium
```

```python
m = folium.Map(location=[34.2,-118.2], 
               zoom_start = 9,
               tiles='CartoDB positron', 
               attribution='CartoDB')

# plot chorpleth over the base map
folium.Choropleth(
                  geo_data=tracts1, # geo data
                  data=tracts1, # data          
                  key_on='feature.properties.FIPS', # key, or merge column
                  columns=['FIPS','Total: No Health Insurance Coverage'], # [key, value]
                  fill_color='BuPu',
                  line_weight=0.1, 
                  fill_opacity=0.8,
                  line_opacity=0.2, # line opacity (of the border)
                  legend_name='Population with No Health Insurance').add_to(m)   
m
```

```python
m = folium.Map(location=[34.2,-118.2], 
               zoom_start = 9,
               tiles='CartoDB positron', 
               attribution='CartoDB')

# plot chorpleth over the base map
folium.Choropleth(
                  geo_data=tracts1, # geo data
                  data=tracts1, # data          
                  key_on='feature.properties.FIPS', # key, or merge column
                  columns=['FIPS','Total: With Health Insurance Coverage'], # [key, value]
                  fill_color='BuPu',
                  line_weight=0.1, 
                  fill_opacity=0.8,
                  line_opacity=0.2, # line opacity (of the border)
                  legend_name='Population With Health Insurance').add_to(m)
m
```

I think it would be more telling to understand the percent of people who do not carry insurance out of the population. In order to do so I am going to define a new variable within the dataset tracts1 that takes the total number of people with no health insurance and divdeds that by the total with no health coverage plus the total with health coverage, and then multiplied by 100. 

```python
tracts1['PCT_No Health Insurance'] = tracts1['Total: No Health Insurance Coverage']/(tracts1['Total: No Health Insurance Coverage']+tracts1['Total: With Health Insurance Coverage'])*100
```

I am fist going to just plot it without a basemap to see the numbers

```python
tracts1.plot(figsize=(12,10),
                 column='PCT_No Health Insurance',
                 legend=True, 
                 scheme='equal_interval')
```

And below I layered the % of population with no health insurance on a folium map as well. 

```python
m = folium.Map(location=[34.2,-118.2], 
               zoom_start = 9,
               tiles='CartoDB positron', 
               attribution='CartoDB')

# plot chorpleth over the base map
folium.Choropleth(
                  geo_data=tracts1, # geo data
                  data=tracts1, # data          
                  key_on='feature.properties.FIPS', # key, or merge column
                  columns=['FIPS','PCT_No Health Insurance'], # [key, value]
                  fill_color='BuPu',
                  line_weight=0.1, 
                  fill_opacity=0.8,
                  line_opacity=0.2, # line opacity (of the border)
                  legend_name='% Population with No Health Insurance').add_to(m)  
m
```
